```python
import os
from PIL import Image, ImageOps

folder="static-US26-SF-travel"

# Create web_imgs folder if it doesn't exist
web_imgs_path = f"{folder}/web_imgs"
os.makedirs(web_imgs_path, exist_ok=True)

# Process each image in original_imgs
img_folders = os.listdir(f"{folder}/original_imgs")
for img_file in img_folders:
    img_path = f"{folder}/original_imgs/{img_file}"
    if os.path.isfile(img_path):
        with Image.open(img_path) as img:
            if os.path.isfile(f"{web_imgs_path}/{img_file}"):
                print(f"{web_imgs_path}/{img_file} already exists, skipping...")
                continue
            # Calculate new height maintaining aspect ratio
            original_width, original_height = img.size
            new_width = 640
            new_height = int(original_height * new_width / original_width)
            # Resize the image
            _resized_img = img.resize((new_width, new_height), Image.LANCZOS)
            _resized_img = ImageOps.exif_transpose(_resized_img)
            resized_img = Image.new(_resized_img.mode, _resized_img.size)
            resized_img.putdata(list(_resized_img.get_flattened_data()))
            # Save to web_imgs folder
            resized_img.save(f"{web_imgs_path}/{img_file}")
```
